## Actividad 1: Recorrer un `BST` de mayor a menor

A continuación hemos dejado el código de la clase `BST` adaptado a la representación de un *conjunto*. Para ello hemos eliminado el campo `_value` de la clase `BST._Node` y hemos sustituido el método `__getitem__` por `add`:

````python
class BST:

    class _Node:
        def __init__(self, key, left=None, right=None):
            self._key = key
            self._left = left
            self._right = right
            
    def __init__(self): # create empty binary tree
        self._root = None
        self._size = 0

    def __len__(self):
        return self._size

    def __contains__(self, key):
        node = self._root
        while node is not None:
            if key == node._key:
                return True
            elif key < node._key:
                node = node._left
            else:
                node = node._right
        return False

    def add(self, key):
        prev, node = None, self._root
        while node is not None:
            prev = node
            if key == node._key:
                return
            elif key < node._key:
                node = node._left
            else:
                node = node._right
        newchild = BST._Node(key)
        self._size += 1
        if prev is None:
            self._root = newchild
        elif key < prev._key:
            prev._left = newchild
        else:
            prev._right = newchild

    def _height_aux(self, node):
        if node is None or (node._left is None and
                            node._right is None):
            return 0
        return 1+max(self._height_aux(node._left),
                     self._height_aux(node._right))

    def height(self):
        return self._height_aux(self._root)
 
    def _find_gt_rec(self, node, key):
        if node is not None: # caso árbol no vacío
            if key >= node._key: # buscar en subárbol derecho
                return self._find_gt_rec(node._right, key)
            else: # else es opcional por el return de arriba
                # buscamos en subárbol izquierdo y lo guardamos:
                resul = self._find_gt_rec(node._left, key)
                if resul is not None: # si hay resultado
                    return resul      # lo devolvemos
                else:
                    return node._key
            
    def find_gt(self, key): # sucesor
        return self._find_gt_rec(self._root, key)

    def _inorder_traversal(self, node):
        if node is not None:
            yield from self._inorder_traversal(node._left)
            yield node
            yield from self._inorder_traversal(node._right)
                
    def __iter__(self):
        for node in self._inorder_traversal(self._root):
            yield node._key

    def __repr__(self):
        aux = []
        for node in self._inorder_traversal(self._root):
            aux.append(repr(node._key))
        return "{" + ",".join(aux) + "}"
    
    # COMPLETAR:_
    # en este punto puedes añadir un método auxiliar llamado al_reves_aux
    # para ayudar al método lista_al_reves
    
    def lista_al_reves(self):
        resultado = []
        for i in self:
            resultado.append(i)
        resultado = list(reversed(resultado))
        return resultado
        
````

Se pide completar el método `lista_al_reves` de manera que recorrar todo el árbol desde el mayor hasta el menor de los elementos. Si, por ejemplo, ejecutas el siguiente código:

```python

for i in (30, 10, 20, 5, 15, 25, 40):
    b.add(i)

b.lista_al_reves()
```

El resultado correcto será:

```python
[40, 30, 25, 20, 15, 10, 5]
```

**Nota:** El método se limita a rellenar una lista python. Puedes utilizar `yield` pero no se exige hacerlo así (es posible utilizar directamente un `append` donde corresponda).

````python
b = BST()
for i in (30, 10, 20, 5, 15, 25, 40):
    b.add(i)

b.lista_al_reves()
````

## Actividad 2: Cargar un diccionario como árbol balanceado

En esta segunda actividad se pide cargar todas las palabras del fichero `10000.txt` (las 10000 palabras más frecuentes del castellano, según la [RAE](http://corpus.rae.es/lfrecuencias.html)) y meterlas en un árbol binario de búsqueda.

Para que quede balanceado, se propone crear una clase `BSTce` (`ce` de `c`onstrucción `e`quilibrada) que hereda de `BST` y que redefine el constructor `__init__` para que reciba una lista de valores que luego debe insertar en el árbol de manera que éste quede lo más equilibrado posible.

Para ello, tras ordenar la lista de valores recibida, va insertando en primer lugar el elemento de la posición central (puede utilizar el método `add`) y, de manera recursiva, aplica esa misma idea a cada mitad recursivamente.

Si la construcción equilibrada funciona correctamente, la altura del diccionario resultante será de 13.

````python
class BSTce(BST): # observa que hereda de BST, podemos acceder a sus métodos y atributos
    def __init__(self, datosIniciales=()):
        super().__init__() # llamamos al constructor de BST
        listaOrdenada = sorted(x for x in datosIniciales)
        # COMPLETAR con una llamada a _inserta_lista_ordenada
        self._inserta_lista_ordenada(listaOrdenada, 0, len(listaOrdenada)-1)
        
    def _inserta_lista_ordenada(self, thelist, first, last):
        # realiza inserciones en el propio objeto (llama a self.add) para meter
        # los objetos de thelist comprendidos entre first y last (inclusive)
        
        if last == first:
            self.add(thelist[first])
        
        elif last-first == 1:
            self.add(thelist[first])
            self.add(thelist[last])

        else:
            self.add(thelist[(((last-first)//2)+first)])
            self._inserta_lista_ordenada(thelist, first, ((last-first)//2)-1+first)
            self._inserta_lista_ordenada(thelist, ((last-first)//2)+1+first, last)

        # COMPLETAR método recursivo


palabras = []
with open('10000.txt') as f:
    palabras = f.read().split()

print(palabras)
               
dicci = BSTce(palabras)

# calcular y mostrar el tamaño y la altura del árbol generado:
print(f"Tiene {len(dicci)} nodos y una altura igual a {dicci.height()}")
````

## Actividad 3: Simular la función de autocompletado

En esta actividad debes simular la función de autocompletado que tienen algunas aplicaciones como editores de texto, etc.

Para ello debes añadir un método llamado `autocompletado` en la clase `BSTautocompletado` (que hereda de `BSTce`) que recibe un prefijo `pref` y un número `n` (con valor por defecto igual a `10`) y que devuelve **una lista Python** con hasta `n` palabras (menos si no hay tantas) de manera que sean:

- Si `pref` forma parte del diccionario hay que considerarla una de las candidatas del autocompletado (obviamente, en otro caso no).
- Debemos llegar hasta `n` palabras utilizando las sucesoras de `pref` (la sucesora, la sucesora de la sucesora, ...)
- Todas las palabras candidatas del autocompletado deben tener a `pref` **como prefijo** (en cuanto encuentras una que ya no es prefijo, ninguna sucesora más lo será).

Por ejemplo, si el usuario introduce palabra `'casa'`, el método `autocompletado` debe devolver la lista de las hasta 10 primeras palabras (en orden alfabético) que puedan completar `'casa'` para formar una palabra:

```python
['casa', 'casada', 'casado', 'casarse', 'casas']
```

Otro ejemplo: el resultado de llamar a `autocompletado` con la palabra `'circ'` será:

```python
['circo',
 'circuito',
 'circuitos',
 'circulación',
 'circular',
 'circunstancia',
 'circunstancias']
```

pero si quieres autocompletar `'circo'` sólo será ella misma (el conjunto de palabras no contiene otras que empiecen por el prefijo `circo`):


```python
['circo']
```

Finalmente, si la palabra utilizada no permite completar ninguna palabra (ejemplo: `'xy'`), el resultado será la lista vacía:

```python
[]
```

Os recordamos que las cadenas Python disponen el método `startswith` para determinar si una cadena es prefijo de otra:

````python
print('camarero'.startswith('cama'))
````

````python
print('camarero'.startswith('hola'))
````

````python
class BSTautoc(BSTce):
    def autocompletado(self, prefijo, n=10):
        opciones = []
        if prefijo in self:
            opciones.append(prefijo)
        sucesor = self.find_gt(prefijo)
        if sucesor.startswith(prefijo):
            opciones.append(sucesor)
        else:
            return opciones
        
        seguir = True
        while (len(opciones)<n) and seguir:
            sucesor = self.find_gt(sucesor)
            if sucesor.startswith(prefijo):
                opciones.append(sucesor)
            else:
                seguir = False
        
        return opciones

            
        
with open('10000.txt') as f:
    palabras = f.read().split()
               
dicci = BSTautoc(palabras)
````
